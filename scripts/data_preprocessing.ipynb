```python
import yfinance as yf
# from sklearn import preprocessing
import numpy as np
import pandas as pd
```

```python
# Fetch the historical closing prices within the specified date range

def fin_time_series(ticker, start_date='2021-01-01', end_date='2023-12-31', interval='1d'):
    """ Returns price and time in the format [sequence length, 2] """
    # Fetch historical closing prices
    close_prices = yf.Ticker(ticker).history(start=start_date, end=end_date, interval=interval)['Close']
    
    
    return close_prices
```

```python
# Choose stocks, start and end dates
# These are 79 large cap US stocks that have date between 2005-01-01 and 2020-12-31

tickers = ['AAL', 'AAPL', 'ABT', 'ACN', 'ADBE', 'AMD', 'AMZN', \
        'AXP', 'BA', 'BAC', 'BKNG', 'BLK', 'BMY', 'BP', 'BTI', \
        'C', 'CAT', 'CMCSA', 'COP', 'COST', 'CRM', 'CSCO', 'CVS', \
        'CVX', 'DE', 'DIS', 'F', 'FDX', 'GE', 'GILD', 'GOOGL', 'GS', \
        'HD', 'HMC', 'HON', 'HPQ', 'HSBC', 'IBM', 'INFY', 'INTC', \
        'JNJ', 'JPM', 'KO', 'LLY', 'LMT', 'LOW', 'LUV', 'MCD', 'MDLZ', \
        'MS', 'MSFT', 'MU', 'NKE', 'NVDA', 'ORCL', 'PEP', 'PFE', 'PG', \
        'QCOM', 'RTX', 'RY', 'RYAAY', 'SBUX', 'SCHW', 'SHEL', 'SNY',\
        'T', 'TGT', 'TM', 'TSM', 'TXN', 'UBS', 'UL', 'UNH', 'VZ', \
        'WFC', 'WM', 'WMT', 'XOM']

start_date = "2005-01-01"
end_date = "2020-12-31"
```

```python
# Loop over each ticker, process data, and save to files

df = pd.DataFrame()

# Loop over each ticker, process data, and add to DataFrame
for t in tickers:
    train_data = fin_time_series(ticker=t, start_date=start_date, end_date=end_date)
    df[t] = train_data
    print(f"Processed data for {t}")

# Save the DataFrame to a CSV file
df.to_csv("../data/stocks_data.csv", index=False)
print("Saved DataFrame to CSV.")
```
